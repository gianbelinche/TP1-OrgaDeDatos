```python
import pandas as pd
datos = pd.read_csv("Entrenamieto_ECI_2020.csv")

filtered = datos[['Region', 'Stage']]

filtered['Stage'] = (filtered['Stage'] == 'Closed Won').astype(int)
filtered = filtered.groupby("Region").agg({'Stage' : ['mean', 'count']})
filtered.columns = ['Closed_Won_mean', 'Closed_Won_count']
filtered = filtered.loc[filtered.Closed_Won_count > 20, ['Closed_Won_mean']].sort_values('Closed_Won_mean', ascending=False)
filtered
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
filtered.plot(kind="bar",color=(0,0.9,0.3,0.8), figsize=(16, 8))
style.use("seaborn")
plt.title("Region vs Closed Won Probability", fontdict={'size' : 18})
plt.xlabel("Region", fontdict={'size' : 16})
plt.ylabel("Closed Won Probability", fontdict={'size' : 16})
plt.show()
```
